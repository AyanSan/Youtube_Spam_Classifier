```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import BernoulliNB
df=pd.read_csv(r"C:\Users\Ays\Documents\youtube\Youtube01-Psy.csv") #replace with own path to files
df1 = pd.read_csv(r"C:\Users\Ays\Documents\youtube\Youtube02-KatyPerry.csv") #replace with own path to files
df2 = pd.read_csv(r"C:\Users\Ays\Documents\youtube\Youtube03-LMFAO.csv") #replace with own path to files
df3 = pd.read_csv(r"C:\Users\Ays\Documents\youtube\Youtube04-Eminem.csv") #replace with own path to files
df4 = pd.read_csv(r"C:\Users\Ays\Documents\youtube\Youtube05-Shakira.csv") #replace with own path to files
df = pd.concat((df,df1))
df = pd.concat((df,df2))
df = pd.concat((df,df3))
df = pd.concat((df,df4))
df
```

```python
df=df[["CONTENT","CLASS"]]
df.head()
```

```python
df["CLASS"]=df["CLASS"].map({0: "Not Spam",
                                   1: "Spam Comment"})
```

```python
df.head()
```

```python
x = np.array(df["CONTENT"])
y = np.array(df["CLASS"])

cv = CountVectorizer()
x = cv.fit_transform(x)
xtrain, xtest, ytrain, ytest = train_test_split(x, y, 
                                                test_size=0.2, 
                                                random_state=42)

model = BernoulliNB()
model.fit(xtrain, ytrain)
print(model.score(xtest, ytest))
```

```python
sample = "Click here for a prize: https://prize.com/" 
data = cv.transform([sample]).toarray()
print(model.predict(data))
```

```python
def preprocess(text):
    new_text = []
    for t in text.split(" "):
        t = '@user' if t.startswith('@') and len(t) > 1 else t
        t = 'http' if t.startswith('http') else t
        new_text.append(t)
    return " ".join(new_text)


def sentiment_hate_texts(texts):
    from tqdm import tqdm
    from transformers import AutoModelForSequenceClassification, AutoTokenizer, AutoConfig
    import numpy as np
    from scipy.special import softmax

    sentiment_score = 0
    hateful_comments = []
    
    sentiment_MODEL = "cardiffnlp/twitter-roberta-base-sentiment-latest"
    sentiment_tokenizer = AutoTokenizer.from_pretrained(sentiment_MODEL)
    sentiment_config = AutoConfig.from_pretrained(sentiment_MODEL)
    sentiment_model = AutoModelForSequenceClassification.from_pretrained(sentiment_MODEL)

    hate_MODEL = "cardiffnlp/twitter-roberta-base-hate-latest"
    hate_tokenizer = AutoTokenizer.from_pretrained(hate_MODEL)
    hate_config = AutoConfig.from_pretrained(hate_MODEL)
    hate_model = AutoModelForSequenceClassification.from_pretrained(hate_MODEL)

    for text in tqdm(texts):
        text = preprocess(text)

        sentiment_encoded_input = sentiment_tokenizer(text, return_tensors='pt')
        sentiment_output = sentiment_model(**sentiment_encoded_input)
        sentiment_scores = sentiment_output[0][0].detach().numpy()
        sentiment_scores = softmax(sentiment_scores)

        sentiment_ranking = np.argsort(sentiment_scores)[::-1]

        hate_encoded_input = hate_tokenizer(text, return_tensors='pt')
        hate_output = hate_model(**hate_encoded_input)
        hate_scores = hate_output[0][0].detach().numpy()
        hate_scores = softmax(hate_scores)

        hate_ranking = np.argsort(hate_scores)[::-1]

        #print(f"\n{text}\n")
        
        
        for i in range(sentiment_scores.shape[0]):
            l = sentiment_config.id2label[sentiment_ranking[i]]
            s = sentiment_scores[sentiment_ranking[i]]
            #print(f"{i+1}) {l} {np.round(float(s), 4)}")
        
       
        if sentiment_ranking[0] == 2:
            sentiment_score += 1
        elif sentiment_ranking[0] == 0:
            sentiment_score -= 1
        else:
            continue

        #print("\n")

       
        for i in range(hate_scores.shape[0]):
            l = hate_config.id2label[hate_ranking[i]]
            s = hate_scores[hate_ranking[i]]
            #print(f"{i+1}) {l} {np.round(float(s), 4)}")

        
        if hate_ranking[0] != 0:  
            hateful_comments.append(text)

    print(f"Average Sentiment Score: {sentiment_score / len(texts)}")
    print(f"Hateful Comments: {hateful_comments}")
    return hateful_comments
```
